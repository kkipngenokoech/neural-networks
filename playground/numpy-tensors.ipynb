CONVERSION FROM NUMPY ARRAYS TO TENSORS AND VICE-VERSA

tensors are numpy arrays that are capable of being executed on GPU, if you create a tensor from numpy array, and vice versa, they both share the same underlying memory

```python
import numpy as np
import torch
```

```python
tensor = torch.rand(3,4)
tensor.type
```

```python
tensor.numpy() #converting a tensor to a numpy array
```

```python
type(tensor.numpy())
```

```python
numpy_arr = np.array([[1.0, 2.0, 4.0],[1.2,3.9,5.6],[100.12, 123.23,143.54]])
numpy_arr
```

```python
tensor_from_ndarray = torch.from_numpy(numpy_arr)
tensor_from_ndarray
```

```python
type(tensor_from_ndarray)
```

```python
torch.is_tensor(tensor_from_ndarray)
```
