```python
import tensorflow as tf
```

A representation of layer = tf.keras.layers.Dense(units=R) (where R is the number of neurons) manually is shown below:

```python

class MyDenseLayer(tf.keras.layers.Layer):
    def __init__(self, input_dim, output_dim):
        super(MyDenseLayer, self).__init__()
        self.W = self.add_weight([input_dim, output_dim])
        self.b = self.add_weight([1, output_dim])
    
    def call(self, inputs):
        # Forward propagation of the inputs
        Z = tf.matmul(inputs, self.W) + self.b
        
        # Feed through a non-linear layer
        output = tf.math.sigmoid(Z)
        
        return output
```

A representation of the above network in tensorflow is:

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(4), # hidden layer with n neurons
    tf.keras.layers.Dense(2), # output layer with two neurons
])
```
