To add and remove dimensions,we can use squeeze and unsqueezed random variables

```python
twobyfour_tensor = torch.rand(2,4)
print(twobyfour_tensor)
```

```python
unsqueezed = torch.unsqueeze(twobyfour_tensor, 2)# add dimensions at index 2, inner most 
print(twobyfour_tensor)
unsqueezed.size()
```

```python
sorted_tensors, sorted_indices = torch.sort(twobyfour_tensor)
sorted_tensors, sorted_indices
```

```python
tensor_float = torch.FloatTensor([-1.1, -2.2, 3.3])
print(torch.unsqueeze(tensor_float, 0))
print(torch.squeeze(tensor_float, 0))
tensor_float
```

```python
torch.abs(tensor_float) # absolute value
```

```python
rand1 = torch.rand(2,3)
print(rand1)
rand2 = torch.rand([2,3])
rand2
```

```python
print(rand1 + rand2) # element wise operations
print(torch.add(rand1, rand2))
print(torch.abs(rand1 - rand2))
print(torch.sub(rand1, rand2))
print(torch.div(rand1, rand2))
print(torch.mul(rand2, rand1))
print(torch.clamp(rand2, min = 1.2, max = 2)) # constrains your tensors to specific range - min and max range values
```

```python
t1 = torch.rand(2,3,5)
t3 = torch.rand(2,3,5)
t2 = torch.rand(5,3,2)
print(torch.mul(t1, t3))
print(torch.dot(torch.Tensor([1,2,3]), torch.Tensor([4,5,6]))) # sums them up to a scalar
```
